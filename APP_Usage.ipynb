```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
from faker import Faker
from tqdm.notebook import tqdm
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = {
    'session_id':[],
    'user_id':[],
    'date':[],
    'time':[],
    'age':[],
    'gender':[],
    'location':[],
    'retention_time':[],
    'app_crash':[],
    'device_type':[],
    'language':[],
    'f1_transfer':[],
    'f2_bill_payment':[],
    'f3_check_balance':[],
    'f4_create_FD_acc':[],
    'user_rate':[],
}
```

```python
fake = Faker()
```

```python
def generate_random_date(start_date, end_date):
    start_datetime = datetime.strptime(start_date, "%Y-%m-%d")
    end_datetime = datetime.strptime(end_date, "%Y-%m-%d")

    # Calculate the difference in days between start and end dates
    days_difference = (end_datetime - start_datetime).days
    random_days = random.randint(0, days_difference)
    random_date = start_datetime + timedelta(days=random_days)

    return random_date.strftime("%Y-%m-%d")
```

```python
import random

def generate_random_time():
    probabilities = [0.05, 0.2, 0.35, 0.35, 0.05]
    time_ranges = [(0, 5), (6, 10), (11, 15), (16, 20), (21, 23)]
    hour_range = random.choices(range(len(time_ranges)), weights=probabilities)[0]
    hour = random.randint(time_ranges[hour_range][0], time_ranges[hour_range][1])
    minute = random.randint(0, 59)

    return f"{hour:02d}:{minute:02d}"
```

```python
# Initialize an empty set to keep track of generated user IDs
generated_user_ids = set()

def generate_unique_user_id():
    while True:
        user_id = np.random.randint(1000, 50000)
        if user_id not in generated_user_ids:
            generated_user_ids.add(user_id)
            return user_id
```

```python
def generate_gender():
    fake = Faker()
    prob_male = 0.7
    prob_female = 0.3

    weighted_genders = [('Male', prob_male), ('Female', prob_female)]
    gender = fake.random_element(elements=weighted_genders)

    return gender
```

```python
def generate_age():
    fake = Faker()

    prob_18_30 = 0.6
    prob_31_40 = 0.3
    prob_41_50 = 0.1

    weighted_age_ranges = [(random.randint(18, 30), prob_18_30),
                           (random.randint(31, 40), prob_31_40),
                           (random.randint(41, 50), prob_41_50)]
    age = fake.random_element(elements=weighted_age_ranges)[0]

    return age
```

```python
no_records = 12000
start_date = "2024-01-01"
end_date = "2024-03-17"

start_session_id = 1000

for session_id in tqdm(range(start_session_id, start_session_id + no_records), desc="Processing session"):  #(not change)
    print("ITERATION OUTER LOOP: ", session_id)
    user_id = generate_unique_user_id()
    location = np.random.choice(['Central Province', 'Eastern Province', 'Northern Province', 'North Central Province', 'Western Province', 'Sabaragamuwa Province', 'Southern Province', 'Uva Province','Central Province', 'Western Province','North Western Province'])

    prob_android = 0.7
    prob_ios = 0.3

    device_type = np.random.choice(['Android','IOS'], p=[prob_android,prob_ios])

    prob_eng = 0.8
    prob_sin = 0.15
    prob_tam = 0.05

    language = np.random.choice(['English','Sinhala','Tamil'], p=[prob_eng,prob_sin,prob_tam])
    no_of_session = np.random.randint(2,30)

    prob_male = 0.7
    prob_female = 0.3
    gender = np.random.choice(['Male', 'Female'], p=[prob_male,prob_female])

    probabilities = [0.05, 0.15, 0.25, 0.35, 0.2]
    user_rate = np.random.choice(range(1, 6), p=probabilities)

    age = generate_age()

    for log in range(1, no_of_session + 1):
      print("ITERATION INNER LOOP:", log)

      data['session_id'].append(session_id)
      data['user_id'].append(user_id)
      data['location'].append(location)
      data['device_type'].append(device_type)
      data['language'].append(language)
      data['gender'].append(gender)
      #data['dob'].append(dob)
      data['age'].append(age)
      data['user_rate'].append(user_rate)

      date = generate_random_date(start_date, end_date)
      data['date'].append(date)

      time = generate_random_time()
      data['time'].append(time)

      prob_no = 0.95
      prob_yes = 0.05

      app_crash = np.random.choice(['No','Yes'], p=[prob_no, prob_yes])
      data['app_crash'].append(app_crash)


      prob_0_f1 = 0.25
      prob_1_f1 = 0.75
      f1_transfer = random.choices([0, 1], weights=[prob_0_f1, prob_1_f1])[0]
      data['f1_transfer'].append(f1_transfer)

      prob_0_f2 = 0.5
      prob_1_f2 = 0.5
      f2_bill_payment = random.choices([0, 1], weights=[prob_0_f2, prob_1_f2])[0]
      data['f2_bill_payment'].append(f2_bill_payment)

      prob_0_f3 = 0.05
      prob_1_f3 = 0.95
      f3_check_balance = random.choices([0, 1], weights=[prob_0_f3, prob_1_f3])[0]
      data['f3_check_balance'].append(f3_check_balance)

      prob_0_f4 = 0.95
      prob_1_f4 = 0.05
      f4_card_payment = random.choices([0, 1], weights=[prob_0_f4, prob_1_f4])[0]
      data['f4_create_FD_acc'].append(f4_card_payment)

      retention_time = np.random.randint(2,10)
      data['retention_time'].append(retention_time)

```

```python
 for key, value in data.items():
    print(f'Length of {key}: {len(value)}')
```
